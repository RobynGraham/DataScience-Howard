# 5. Which states have the highest / lowest acceptance rates

```python
hbcus_2018['State'].unique().size
```

```python
""" Algorithm

1. Check each row in the table

2. For each state, caluculate the average acceptance rate

3. Create plots

 """

def get_averages_visualization(table):

  states = list(table['State'].unique())

  state_list = [] # create this to add to a dataframe later on

  for state in states:
    state_table = hbcus_2018[table['State'] == state ] # Get the state

    state_acceptance_rate_average = round(np.mean(state_table['Acceptance_rate'])) # Get the avg

    num_of_schools = state_table.shape[0] # Get the num of schools

    state_info = [state,state_acceptance_rate_average, num_of_schools] # Put the values in a list ' ['AL', 23.2 ]

    state_list.append(state_info)

  state_table = pd.DataFrame(state_list,columns=['State','Avg_acceptance_rate',"num_of_schools"])

  state_table = state_table.sort_values(by='Avg_acceptance_rate',ascending = False)


  return state_table

state_averages_table = get_averages_visualization(hbcus_2018)

sns.barplot(data=state_averages_table, x="Avg_acceptance_rate", y="State").set(title = "Average Acceptance rate for each school") # Create a visualization


```

```python
num_of_schools = state_averages_table.sort_values(by='num_of_schools')
sns.barplot(data=num_of_schools, x="num_of_schools", y="State").set(title = "Num of schools for each state") # Create a visualization

```
